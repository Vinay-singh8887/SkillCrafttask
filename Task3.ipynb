Build a tool that assesses the strength of a password based on criteria such as length,presence of uppercase and lowercase letters,numbers, and special character

```python
import re

min_length = 8

def check_password_strength(password):

    missing_criteria = []


    if len(password) < min_length:
        missing_criteria.append(f"- Must be at least {min_length} characters long")

    if not re.search(r'[A-Z]', password):
        missing_criteria.append("- Must contain at least one uppercase letter")

    if not re.search(r'[a-z]', password):
        missing_criteria.append("- Must contain at least one lowercase letter")

    if not re.search(r'[0-9]', password):
        missing_criteria.append("- Must contain at least one number (digit)")

    special_chars = r'[!@#$%^&*()_+=\-{}[\].?/\\|`~]'
    if not re.search(special_chars, password):
        missing_criteria.append("- Must contain at least one special character (e.g., !@#$)")

    if not missing_criteria:
        return "Strong ðŸ’ª", []
    else:
        return "Weak âš ", missing_criteria


my_password = input("Enter a password to check: ")
strength, issues = check_password_strength(my_password)

print(f"\nPassword Strength: *{strength}*")

if issues:
    print("\nTo improve your password, you should add:")
    for issue in issues:
        print(issue)
else:
    print("Your password meets all complexity requirements.")
```
