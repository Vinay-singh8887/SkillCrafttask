Create a basic keylogger program that records and logs keystrokes. focus on logging the keys pressed and logging the keys pressed and saving them to a file.

```python
import logging

def setup_logger():
    logging.basicConfig(
        filename='logs/app.log',
        filemode='a',
        format='%(asctime)s - %(message)s',
        level=logging.INFO
    )
    return logging.getLogger()

def main():
    logger = setup_logger()
    print("Enter text (type 'exit' to quit):")
    while True:
        user_input = input("> ")
        if user_input.lower() == "exit":
            break
        logger.info(f"User entered: {user_input}")

if __name__ == "__main__":
    main()
```
