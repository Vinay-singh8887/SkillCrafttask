Create a program that can encrypt and decrypt text using the caesar cipher algorithm.Allow users to input a message and a shift value to perform encryption and decryption

```python
def caesar_cipher(text, shift, mode='encrypt'):

    result = ""

    if mode == 'decrypt':
        shift = -shift

    for char in text:
        if char.isalpha():
            start = ord('A') if char.isupper() else ord('a')
            char_index = ord(char) - start
            new_index = (char_index + shift) % 26
            new_char = chr(start + new_index)
            result += new_char
        elif char.isdigit():
            result += char
        else:
            result += char
    return result

def run_cipher_program():
    print("--- Caesar Cipher Tool ---")

    while True:
        choice = input("Do you want to (E)ncrypt or (D)ecrypt? ").strip().lower()
        if choice in ['e', 'd']:
            mode = 'encrypt' if choice == 'e' else 'decrypt'
            break

        print("Invalid choice. Please enter 'E' or 'D'.")
    message = input("Enter the message: ").strip()

    while True:
        try:
            shift = int(input("Enter the shift value (e.g., 3): ").strip())
            break
        except ValueError:
            print("Invalid shift value. Please enter an integer.")

    final_message = caesar_cipher(message, shift, mode)
    print("-" * 30)
    print(f"Mode: {'Encryption' if mode == 'encrypt' else 'Decryption'}")
    print(f"Shift: {shift}")
    print(f"Original Message: {message}")
    print(f"Processed Message: {final_message}")
    print("-" * 30)

run_cipher_program()
```
